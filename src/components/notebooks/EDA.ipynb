# Exploratory Data Analysis
## Diamond Price Prediction

### Data Description
#### There are 10 independent variables:

    id: Identification column.

    cut: Describe the cut quality of the cubic zirconia. Quality in increasing order: Fair, Good, Very Good, Premium, Ideal.

    color: Colour of the cubic zirconia, with D being the best and J the worst.

    clarity: Clarity refers to the absence of the Inclusions and Blemishes. (In order from Best to Worst, FL = flawless, I3= level 3 inclusions) FL, IF, VVS1, VVS2, VS1, VS2, SI1, SI2, I1, I2, I3

    carat: Carat weight of the cubic zirconia.
    
    depth: The Height of a cubic zirconia, measured from the Culet to the table, divided by its average Girdle Diameter.

    table: The Width of the cubic zirconia's Table expressed as a Percentage of its Average Diameter.

    X: Length of the cubic zirconia in mm.

    Y: Width of the cubic zirconia in mm.

    Z: Height of the cubic zirconia in mm.

    Target variable:

    price: Price of the Diamond.

```python
# Removing unnecessary ID column
df = df.drop(labels=['id'],axis=1)
df.head()
```

```python
# Segregating Numerical and Categorical Feautures
numerical_cols = df.columns[df.dtypes!='object']
categorical_cols = df.columns[df.dtypes=='object']
print("Numerical Features:",numerical_cols.values)
print("Categorical Features:",categorical_cols.values)
```

```python
df['clarity'].value_counts()
```

```python
fig, ax = plt.subplots(ncols=3, nrows=3, figsize=(15,15))
index = 0
ax = ax.flatten()

for col in numerical_cols:
    pl = sns.histplot(data=df,x=col,kde=True,color='paleturquoise',ax=ax[index])
    pl.lines[0].set_color('rosybrown')
    index += 1
    if col == 'price':
        break
    
fig.suptitle(f'Histogram plots of numerical features\n', ha='center',  fontweight='bold', fontsize=20)
plt.tight_layout()
```

**Observation:** 

```python
for col in categorical_cols:
    sns.catplot(data=df, x=col, kind='count', hue=col, aspect=2, palette='viridis',order=df[col].value_counts().index)
    plt.title(f'Count plot for {col}',size=24)
    plt.show()
```

**Observation:** There are some outliers in the dataset.

### **Correlation**

```python
sns.heatmap(data=df[numerical_cols].corr(),annot=True)
```

**Observation:** (X,Y,Z) are highly correlated, so we can drop them.

```python
target = "price"
index, rows, columns = 0,3,2

fig, axs = plt.subplots(nrows=rows, ncols=columns,figsize=(12,12))
for row in range(rows):
    for col in range(columns):
        if numerical_cols[index] == target:
            pass
        
        sns.scatterplot(x = numerical_cols[index], y = target, data = df, ax=axs[row][col])
        index += 1
```

**Observation:** Price depends most on carat than on x and than on y

```python
index, rows = 0,3
colors = ['mediumseagreen', 'mediumturquoise', 'rosybrown']  # Add more colors if you have more rows
fig, axs = plt.subplots(nrows=rows, figsize=(12,12))
for row in range(rows):
    sns.violinplot(x = categorical_cols[index], y=target, data=df,  ax=axs[row], color=colors[row])
    index += 1
```

```python
# Dropping Highly correlated columns
df = df.drop(labels=['x','y','z'],axis=1)
```

```python
cut_map  = {'Fair':1,'Good':2,'Very Good':3,'Premium':4,'Ideal':5}
```

```python
df['clarity'].unique()
```

```python
clarity_map = {'I1':1,'SI2':2,'SI1':3,'VS2':4,'VS1':5,'VVS2':6,'VVS1':7,'IF':8}
```

```python
df['color'].unique()
```

```python
color_map  = {'D':7,'E':6,'F':5,'G':4,'H':3,'I':2,'J':1}
```

```python
df['cut'] = df['cut'].map(cut_map)
df['clarity'] = df['clarity'].map(clarity_map)
df['color'] = df['color'].map(color_map)
```
